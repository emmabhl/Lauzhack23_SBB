```python
##IN EACH TRIP GET THE STOP POINTS/LOCATIONS
def get_nb_legs(journey):
    res = []
    for trip in journey['trips']:
        numberStops = 0
        for leg in trip['legs']:
            for stop_point in leg['serviceJourney']['stopPoints']:
                if stop_point['place']['type'] == 'StopPlace':
                    numberStops += 1 
        res.append({'id' : trip['id'], 'numberLegs' : len(trip['legs']), 'TotNumberStops': numberStops})
    return res
```
